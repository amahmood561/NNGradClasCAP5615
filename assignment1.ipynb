#### Report all samples with respect to the Crim index on a plot (the x-axis show the index of the sample, and the y-axis shows the crim index of the sample).

```python
def CrimIndexOnPlotReq():
    dataframecrim = df['Crim']
    x=np.arange(dataframecrim.shape[0]) 
    plt.figure()
    plt.figure(figsize=(8,5))
    plt.scatter(x, dataframecrim, marker='+')
    plt.title('Crim Index')
    plt.xlabel('The index of the sample')
    plt.ylabel('The crim index of the sample')
    plt.show()
CrimIndexOnPlotReq()
```

#### Show both histogram of the Crim index and the density of the Crim index on a 1x2 frame.

```python
import seaborn as sns
def histogramCrimIndexAndDensity():
    plt.figure(figsize=(22,5))
    plt.subplot(1,2,1)
    sns.distplot(df['Crim'], hist=True, kde=False, bins=20, color = 'black',hist_kws={'edgecolor':'black'})
    plt.title('Crim Histogram')
    plt.subplot(1,2,2)
    sns.distplot(df['Crim'], hist=True, kde=True, bins=25, color = 'red',hist_kws={'edgecolor':'black'},kde_kws={'linewidth':2})
    plt.title('Density Crim Index')
    plt.show()
histogramCrimIndexAndDensity()
```

#### Show following four scatter plots in one frame (1x4), crim-medv, Rm-medv, Agemedv,Tax-medv, and explain how are they (Crim, Rm, Age, Tax) correlated to the medium house value (Medv)


```python
def scatterplotsmedvcoorelation():
    plt.figure(figsize=(20, 5))
    selectedAttributes = ['Crim', 'Rm','Age','Tax']
    medv = df['Medv']
    for i, col in enumerate(selectedAttributes):
        plt.subplot(1, len(selectedAttributes) , i+1)
        x = df[col]
        y = medv
        plt.scatter(x, y, marker='+')
        plt.title(col)
        plt.xlabel(col)
        plt.ylabel('Medv')
scatterplotsmedvcoorelation()
```

#### Create a subset which only includes properties with Crim less than 1 (inclusive), and Rm greater than 6 (inclusive).


```python
def subset():
    #Create a subset which only includes properties with Crim less than 1 (inclusive), and Rm 
    #greater than 6 (inclusive).
    df.filter = df[(df['Crim']<=1) & (df['Rm']>=6)] 
    arrangedfiltershape=np.arange(df.filter.shape[0]) 
    plt.figure(figsize=(8, 6))
    plt.scatter(arrangedfiltershape, df.filter['Rm'], marker='+')
subset()
```

#### Show a scatter plot between Rm and Medv (x-axis shows Rm and y-axis denote Medv), please color all properties with 7 or larger Rm values as red, and rest properties as black.

```python
import numpy as np
import matplotlib.pyplot as plt
def scatterrmandmedv():
    rm= df['Rm']
    medv = df['Medv']
    columns=[]
    for i in range(0,len(rm)):
        columns.append('black') if rm[i]<=7 else columns.append('red') 
    plt.figure(figsize=(8, 7))
    for i in range(len(rm)):
        plt.scatter(rm[i],medv[i],c=columns[i],s=10, linewidth=0)
        plt.xlabel('Rm')
        plt.ylabel('Medv')
        plt.title('Scatter plot requirement of Rm and Medv')
scatterrmandmedv()
```

#### Read the movie100.csv, and check average length (in terms of number of words) of the positive and negative reviews, respectively

```python
#read the review100
reviewdf = pd.read_csv("review100.csv") 
postiveSentimentReviews = reviewdf[reviewdf['sentiment'].str.contains('positive')]
positive_string_length_for_each_row = postiveSentimentReviews['movie review'].str.len()
print("average review length:" + str( positive_string_length_for_each_row.mean()))

negativeSentimentReviews = reviewdf[reviewdf['sentiment'].str.contains('negative')]
negative_string_length_for_each_row = negativeSentimentReviews['movie review'].str.len()
print("average review length:" + str( negative_string_length_for_each_row.mean()))
```

#### Check the frequency of the words with respect to positive and negative reviews, respectively. Print the top 20 most frequent words (including words and their frequency), for positive and negative reviews, respectively 

```python
postiveTop20WordCount = postiveSentimentReviews['movie review'].str.split(expand=True).stack().value_counts().head(20)
print(postiveTop20WordCount)
print("##################################")
negativeTop20WordCount = negativeSentimentReviews['movie review'].str.split(expand=True).stack().value_counts().head(20)
print(negativeTop20WordCount)
```

#### Draw a bar plot to show the frequency of the two-20 most frequent words (the x-axis denotes words, and the y-axis denotes their frequency), for positive and negative reviews, respectively

```python
import matplotlib.pyplot as plt

def createNegativeTop20Bar():
    indexWords = negativeTop20WordCount.index.tolist()
    counts = negativeTop20WordCount.tolist()
    figure = plt.figure()
    graph = figure.add_axes([0,0,1,1])
    graph.bar(indexWords,counts)
    plt.show()
createNegativeTop20Bar()
```

```python
import matplotlib.pyplot as plt

def createPositiveTop20Bar():
    indexWords = postiveTop20WordCount.index.tolist()
    counts = postiveTop20WordCount.tolist()
    figure = plt.figure()
    graph = figure.add_axes([0,0,1,1])
    graph.bar(indexWords,counts)
    plt.show()
createNegativeTop20Bar()
```
