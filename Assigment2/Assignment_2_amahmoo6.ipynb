#### • Please use 80% of instances in the “housing.header.binary.txt” dataset to build a decision 
#### tree classifier (using all features) to predict house value Medv [0.5 pt]. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
from sklearn.metrics import confusion_matrix 
from sklearn import metrics
from sklearn.metrics import roc_auc_score 
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split
```

```python
q8dataframe = pd.read_csv("housing.header.binary.txt") 
q8dataframe.head()
```

```python
feature = q8dataframe.iloc[:,0:-1]
lab = q8dataframe.loc[:,['Medv']]
feature
```

```python
lab
```

```python
from sklearn.model_selection import train_test_split
trainA, testA, trainB, testB = train_test_split(feature,lab,test_size=.2,random_state=0)
print(trainA.shape,testA.shape,trainB.shape,testB.shape) 
```

```python
q8_decision_tree_object = tree.DecisionTreeClassifier() 
q8_decision_tree_object.fit(trainA, trainB)
ppredictionforb= q8_decision_tree_object.predict(testA) 
print(testB,ppredictionforb)
```

```python
plt.rcParams.update({'font.size':17})
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (5,2), dpi=500)
tree.plot_tree(q8_decision_tree_object,feature_names=feature.columns, class_names =['0','1 '],filled = True)
```

```python
q8_confusion_matrix = confusion_matrix(testB, ppredictionforb)
q8_confusion_matrix
```

```python
def confusionmatrix(q8_confusion_matrix):
    true_negative, false_postivie, false_negative, true_positive = q8_confusion_matrix.ravel() 
    print("*********************************************************")
    print('true negative',true_negative)
    print('false positive =',false_postivie)
    print('False negatuve =',false_negative)
    print('True positive =',true_positive)
    print("*********************************************************")
    accuracy = float(true_negative+true_positive)/float(true_negative+false_postivie+false_negative+true_positive)
    true_postive_rate = true_positive/(true_positive + false_negative)
    false_postive_rate = false_postivie/(false_postivie + true_negative)
    print("Accuracy Value= %f" % accuracy)
    print("Treu Postive Rate= %f" % true_postive_rate)
    print("False Positive Raye = %f" % false_postive_rate)
    print("*********************************************************")
```

```python
confusionmatrix(q8_confusion_matrix)
```

```python
acc_score = accuracy_score(testB, ppredictionforb)
print("acc_score = " + str(acc_score))
```

```python
probability_value_calced = q8_decision_tree_object.predict_proba(testA) 
print(probability_value_calced[:10])
```

```python
score_for_the_auc= roc_auc_score(testB,probability_value_calced[:,1])
print("AUC calculated value=" + str(score_for_the_auc))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(testB, probability_value_calced[:,1], pos_label=1) 
print(fpr,tpr,thresholds)
```

```python
def AucGraphCreator():
    plt.figure()
    lineWidth = 1
    plt.plot(fpr, tpr, color='darkRed',lw=lineWidth, 
             label='ROC graph auc score = %0.9f' 
             %score_for_the_auc)
    plt.xlabel('False Positive Rate')
    plt.xlabel(' False Positive Rate ')
    plt.ylabel(' True Positive Rate ')
    plt.title("ROC CURVE")
    plt.plot([0, 1], [0, 1], 
             color='purple', 
             lw=lineWidth, 
             linestyle='-') 
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.legend(loc='best')
    plt.show()
AucGraphCreator()
```
